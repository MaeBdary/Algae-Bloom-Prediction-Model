```python
from pathlib import Path
NOTEBOOKS_DIR = Path.cwd().resolve()
MODEL_DIR = (NOTEBOOKS_DIR.parent / "modelsave")
MODEL_DIR.mkdir(parents=True, exist_ok=True)

CSV_PATH = '../Data/final_dataset_synthetic.csv'

```

```python
import re, numpy as np, pandas as pd, joblib, warnings
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

warnings.filterwarnings("ignore")
RNG = 42

df = pd.read_csv(CSV_PATH)

def infer_target(cols):
    pats = [r"phycocyanin", r"\bpc(_?conc|_?concentration)?\b", r"\bpc\b"]
    for p in pats:
        hit = [c for c in cols if re.search(p, c, flags=re.I)]
        if hit:
            hit_sorted = sorted(hit, key=lambda c: 0 if re.search("conc|concentration", c, re.I) else 1)
            return hit_sorted[0]
    raise ValueError("Could not infer target column for phycocyanin.")

target = infer_target(df.columns)

leak = re.compile(r"phyco|phycocyanin|\bpc\b", re.I)
X_cols = [c for c in df.columns if c != target and not leak.search(c)]
X_raw = df[X_cols].copy()
y = df[target].values

num_cols = [c for c in X_raw.columns if pd.api.types.is_numeric_dtype(X_raw[c])]
cat_cols = [c for c in X_raw.columns if c not in num_cols]

preprocess_linear = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")), ("sc", StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")), ("oh", OneHotEncoder(handle_unknown="ignore"))]), cat_cols)
])

preprocess_tree = ColumnTransformer([
    ("num", SimpleImputer(strategy="median"), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")), ("oh", OneHotEncoder(handle_unknown="ignore", min_frequency=10))]), cat_cols)
])
```

```python
from math import sqrt

def cv_report(y_true, y_pred, label="Model"):
    r2  = r2_score(y_true, y_pred)
    rmse = sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    print(f"{label} — CV R²: {r2: .4f} | RMSE: {rmse: .4f} | MAE: {mae: .4f}")

cv = KFold(n_splits=5, shuffle=True, random_state=RNG)
```

```python
from sklearn.linear_model import ElasticNet


model1 = Pipeline([
    ("prep", preprocess_linear),
    ("est",  ElasticNet(alpha=0.05, l1_ratio=0.2, max_iter=5000, random_state=RNG))
])

yhat1 = cross_val_predict(model1, X_raw, y, cv=cv, n_jobs=-1)
cv_report(y, yhat1, "ElasticNet")

model1.fit(X_raw, y)
joblib.dump(model1, MODEL_DIR / "model_elasticnet.joblib")
print(f"Saved: {MODEL_DIR / 'model_elasticnet.joblib'}")
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

model2 = Pipeline([
    ("prep", preprocess_tree),
    ("est",  HistGradientBoostingRegressor(
        learning_rate=0.06,
        max_depth=6,
        max_bins=255,
        min_samples_leaf=10,
        l2_regularization=0.0,
        early_stopping=True,
        random_state=RNG
    ))
])

yhat2 = cross_val_predict(model2, X_raw, y, cv=cv, n_jobs=-1)
cv_report(y, yhat2, "HGBR")

model2.fit(X_raw, y)
joblib.dump(model2, MODEL_DIR / "model_hgbr.joblib")
print(f"Saved: {MODEL_DIR / 'model_hgbr.joblib'}")
```

Using an unsupervised PCA step to extract rough latent factors -> fed into an XGboost Regressor

```python
import numpy as np, pandas as pd, joblib
from math import sqrt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold, cross_val_predict, train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor

num_cols = [c for c in X_raw.columns if pd.api.types.is_numeric_dtype(X_raw[c])]
cat_cols = [c for c in X_raw.columns if c not in num_cols]
pca_components = min(50, max(5, len(num_cols)//2))

preprocess = ColumnTransformer([
    ("num", Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("sc", StandardScaler()),
        ("pca", PCA(n_components=pca_components, svd_solver="randomized", random_state=RNG))
    ]), num_cols),
    ("cat", Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("oh", OneHotEncoder(handle_unknown="ignore"))
    ]), cat_cols)
], sparse_threshold=1.0)

cv = KFold(n_splits=5, shuffle=True, random_state=RNG)
model_cv = Pipeline([
    ("prep", preprocess),
    ("est", XGBRegressor(
        objective="reg:squarederror",
        eval_metric="rmse",
        n_estimators=350,
        learning_rate=0.07,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=RNG,
        tree_method="hist",
        n_jobs=1,
        verbosity=0
    ))
])

yhat_cv = cross_val_predict(model_cv, X_raw, y, cv=cv, n_jobs=4, pre_dispatch=4)
print("PCA+XGB (fast CV) — R²: {:.4f} | RMSE: {:.4f} | MAE: {:.4f}".format(
    r2_score(y, yhat_cv),
    sqrt(mean_squared_error(y, yhat_cv)),
    mean_absolute_error(y, yhat_cv)
))

X_tr, X_val, y_tr, y_val = train_test_split(X_raw, y, test_size=0.15, random_state=RNG)
pre_fitted = ColumnTransformer([
    ("num", Pipeline([
        ("imp", SimpleImputer(strategy="median")),
        ("sc", StandardScaler()),
        ("pca", PCA(n_components=pca_components, svd_solver="randomized", random_state=RNG))
    ]), num_cols),
    ("cat", Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("oh", OneHotEncoder(handle_unknown="ignore"))
    ]), cat_cols)
], sparse_threshold=1.0).fit(X_tr)

Xtr_t = pre_fitted.transform(X_tr)
Xval_t = pre_fitted.transform(X_val)

xgb_final = XGBRegressor(
    objective="reg:squarederror",
    eval_metric="rmse",
    n_estimators=2000,
    early_stopping_rounds=50,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=RNG,
    tree_method="hist",
    n_jobs=-1,
    verbosity=0
)
xgb_final.fit(Xtr_t, y_tr, eval_set=[(Xval_t, y_val)], verbose=False)

final_pipe = Pipeline([("prep", pre_fitted), ("est", xgb_final)])

pca = final_pipe.named_steps["prep"].named_transformers_["num"].named_steps["pca"]
print(f"PCA kept {pca.n_components_} components.")

outpath = MODEL_DIR / "model_pca_xgb.joblib"
joblib.dump(final_pipe, outpath)
print(f"Saved: {outpath}")
```

```python
from math import sqrt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

def compare_three(y, yhat1, yhat2, yhat3, names=("ElasticNet", "HGBR", "XGB")):
    def metrics(y, yp):
        return (
            r2_score(y, yp),
            sqrt(mean_squared_error(y, yp)),
            mean_absolute_error(y, yp)
        )
    
    r2_1, rmse_1, mae_1 = metrics(y, yhat1)
    r2_2, rmse_2, mae_2 = metrics(y, yhat2)
    r2_3, rmse_3, mae_3 = metrics(y, yhat3)
    
    print(f"{'Model':<12} | {'R²':>8} | {'RMSE':>8} | {'MAE':>8}")
    print("-"*45)
    print(f"{names[0]:<12} | {r2_1:8.4f} | {rmse_1:8.4f} | {mae_1:8.4f}")
    print(f"{names[1]:<12} | {r2_2:8.4f} | {rmse_2:8.4f} | {mae_2:8.4f}")
    print(f"{names[2]:<12} | {r2_3:8.4f} | {rmse_3:8.4f} | {mae_3:8.4f}")

# Call it:
compare_three(y, yhat1, yhat2, yhat_cv, names=("ElasticNet", "HGBR", "PCA+XGB"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

models = {
    "ElasticNet": yhat1,
    "HGBR": yhat2,
    "PCA-XGB": yhat_cv, 
}

all_vals = np.concatenate([y] + [pred for pred in models.values()])
lo, hi = np.min(all_vals), np.max(all_vals)

fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=True)
for ax, (name, preds) in zip(axes, models.items()):
    sns.scatterplot(x=y, y=preds, alpha=0.6, ax=ax)
    ax.plot([lo, hi], [lo, hi], 'r--', linewidth=1)
    ax.set_title(f"{name}: Actual vs Predicted")
    ax.set_xlabel("Actual Phycocyanin")
    ax.set_xlim(lo, hi)
    ax.set_ylim(lo, hi)

axes[0].set_ylabel("Predicted Phycocyanin")
plt.tight_layout()
plt.show()
```
